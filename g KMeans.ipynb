# K-Means Clustering Example

K Means is a unsupervised learning which is used find new patterns or classification which can turn out to useful insights.

### Aim
To define segments of customer profiles in order to derive the facilities and prefered shopping items which can improve customer experience.

### Data
Data is Downloaded from Kaggle from the below link and author is Ioannis Mesionis <br>
https://www.kaggle.com/ioannismesionis/mall-customer-segmentation-unsupervised-learning

### Libraries used

```python
from numpy import random, array
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random, float
%matplotlib inline
```

### Loading the data

```python
df = pd.read_csv("Data/Mall_Customers.csv")
df.head()
    
```

###  Extracting the features
Only the Annual Income and spending score will be used in this example.


```python
data = np.squeeze(df.filter(['Annual Income','Spending Score'], axis=1).as_matrix())
```

### Finding the optimal value of K
The experiment of modelling and fitting will be repeated for K = 3 to K = 6 to find the optimal value of K.

### Scaling the data
Scaling the data in order to have fair weightage for both features.

```python
scaledData=scale(data)
```

### Defining the model
First K = 3 is tried first

```python
model = KMeans(n_clusters=3)
model = model.fit(scaledData)
```

```python
model.labels_.astype(float)
```

### Plotting the data

```python
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

#### Observation
We can have more segments as we can observe in the graph


### Trying K=4


```python
model = KMeans(n_clusters=4)
model = model.fit(scaledData)
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

#### Observation
Still left top segments can be a separte group , so we increase te value of K

### K=5

```python
model = KMeans(n_clusters=5)
model = model.fit(scaledData)
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

#### Observation
Now the data seems to be classifed properly. But just to be sure we will try a K=6.

### K=6

```python
model = KMeans(n_clusters=6)
model = model.fit(scaledData)
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

#### Observation
Now the data seems to be segmented more than required, hence the value K=5 can be accepted.

### Result:

The data given is best clasiffied into **5 segments** which can be used for further processing for imporving customer experience.

### Conclusion:
The KNN method was used into order to segment the customer profiles of a shopping mall.
